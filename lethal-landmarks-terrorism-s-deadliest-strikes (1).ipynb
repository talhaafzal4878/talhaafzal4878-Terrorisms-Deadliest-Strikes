## ***Key Features***
**About Dataset**
This dataset contain terrorist attack data from 1970 to 2023.

It contains data like:

| **Name** | **Definition** |
|------|------------|
| Entity | Name of the country |
| Entity Code | Abbreviation or code representing the country |
| Attack Year | Year of the terrorist attack |
| Total Terrorist Attacks | Total number of terrorist attacks |
| Terrorism Deaths | Total number of deaths due to terrorism |
| Attack Method: Hijacking | Number of terrorist attacks involving hijacking |
| Attack Method: Hostage Taking (Barricade Incident) | Number of terrorist attacks involving hostage-taking (barricade incident) |
| Attack Method: Unarmed Assault | Number of terrorist attacks involving unarmed assault |
| Attack Method: Facility/Infrastructure Attack | Number of terrorist attacks targeting facilities or infrastructure |
| Attack Method: Hostage Taking (Kidnapping) | Number of terrorist attacks involving hostage-taking (kidnapping) |
| Attack Method: Assassination | Number of terrorist attacks involving assassination |
| Attack Method: Armed Assault | Number of terrorist attacks involving armed assault |
| Attack Method: Bombing/Explosion | Number of terrorist attacks involving bombing or explosion |
| Death Age 100+ | Number of victims aged 100 or above |
| Death Age: 51-99 | Number of victims aged 51 to 99 |
| Death Age: 21-50 | Number of victims aged 21 to 50 |
| Death Age: 11-20 | Number of victims aged 11 to 20 |
| Death Age: 6-10 | Number of victims aged 6 to 10 |
| Death Age: 1-5 | Number of victims aged 1 to 5 |
| Terrorist Death Type: Suicide | Number of terrorist deaths by suicide |
| Terrorist Death Type: Killed | Number of terrorist deaths by being killed |

```python
class dataSetanalyzer:
    # we use a constuctor here
    def __init__(self, dataset, heading_color1='#00203FFF', heading_color2='#b8f2df', text_color='white'):
        self.dataset = dataset
        self.heading_color1 = heading_color1
        self.heading_color2 = heading_color2
        self.text_color = text_color
     # default style of every heading
    def style_heading(self, text):
        return f"""
    <div style = "
        font-family : 'Montserrat', sans-serif;
        text-align : 'center';
        background: linear-gradient(to right, {self.heading_color1}, {self.heading_color2});
        color : {self.text_color};
        padding : 15px;
        font-size : 20px;
        font-style : italic;
        font-weight : bold;
        line-height : 1;
        border-radius : 20px 0px 20px 0px;
        margin-bottom : 20px;
        margin-top : 20px;
        border : 3px inset {self.heading_color1}; 
    ">
        {text}
    </div>
    """
    # (i) style of table but in this index is hide 
    def style_table(self, dataset):
        style_df = dataset.style \
        .set_table_styles([
            {'selector': 'th', 'props': [('color', 'white'), ('background-color', '#00203FFF')]},
            {'selector': 'td', 'props': [('padding', '8px'), ('border-bottom', '1px solid #ddd')]}
        ]) \
        .hide(axis = 'index')
        return style_df.to_html()
    
    # (ii) style of table but minor change is that index is visible
    def style_tablee(self, dataset):
        style_df = dataset.style \
        .set_table_styles([
            {'selector': 'th', 'props': [('color', 'white'), ('background-color', '#00203FFF')]},
            {'selector': 'td', 'props': [('padding', '8px'), ('border-bottom', '1px solid #ddd')]}
        ]) 
        
        return style_df.to_html()

    def data_analysis(self, n_top=5):
        # Rest of the code remains the same
        df_heading = self.style_heading('Sample of Dataset rows is given below')
        display(HTML(df_heading))
        display(HTML(self.style_table(self.dataset.sample(5))))
        
        # statisical summary
        summary_heading = self.style_heading('Statistical Summary of Dataset')
        display(HTML(summary_heading))
        display(HTML(self.style_tablee(self.dataset.describe())))
        
        #check for null values
        null_heading = self.style_heading('Null Values and their Percentage in Dataset')
        display(HTML(null_heading))
        null_count = self.dataset.isnull().sum()
        if null_count.sum() == 0 : 
            display(HTML('<p> There is no null values in Dataset </p>'))
        else:
            display(HTML('<h3> Dataset: </h3>'))
            threshold = 0
            missing_values = df.isnull().sum().sort_values(ascending = False)
            missing_values = missing_values[missing_values > threshold].to_frame()
            display(HTML('Missing Values found: '))
            display(HTML(self.style_tablee(missing_values)))
            
        # percentage of missing values
        display(HTML('<p> Percentage(%) of missing values is this: <p>'))
        null_Percentage = missing_values / len(self.dataset) * 100
        null_Percentage = null_Percentage[null_Percentage>0]
        null_Percentage = null_Percentage.round(2)
        display(HTML(self.style_tablee(null_Percentage)))
        
        # check for duplicate
        duplicates_heading = self.style_heading('Duplicate Values in this Dataset')
        display(HTML(duplicates_heading))
        duplicates_count = self.dataset.duplicated().sum()
        display(HTML(f'{duplicates_count} duplicates  found in dataset.'))
        
        # shape of dataset
        shape_heading = self.style_heading('Shape of Dataset')
        display(HTML(shape_heading))
        total_rows, total_columns = self.dataset.shape
        display(HTML(f'In our dataset, we have {total_rows} Rows and {total_columns} Columns.'))
        
        # unique values
        uniqueValues_heading = self.style_heading('Unique Values in Dataset')
        display(HTML(uniqueValues_heading))
        uniqueValues_table = pd.DataFrame({
            'No.' : list(range(1,len(self.dataset.keys())+1)),
            'Column Name': self.dataset.columns,
            'Data Type': [self.dataset[col].dtype for col in self.dataset.columns],
            'Unique Values': [', '.join(map(str, self.dataset[col].unique()[:9])) for col in self.dataset.columns]
        })
        display(HTML(self.style_table(uniqueValues_table)))
        
        
analyzer = dataSetanalyzer(df).data_analysis()
```

## ***Questions & Answers***
### How has the number of terrorist attacks and terrorism-related deaths changed over the years?

```python
number_of_terrorist_attacks = df.groupby('Year')[['Terrorist attacks','Terrorism deaths' ]].sum().reset_index()
number_of_terrorist_attackss = number_of_terrorist_attacks.sort_values(by = 'Terrorist attacks',ascending = False).head(10)
print('Trend of Terrorist Attacks by years')
print(number_of_terrorist_attackss)
# lets visualize it
plt.figure(figsize = (16,5))
background_color = '#5fa1bc'
sns.set_theme(style="whitegrid")
sns.lineplot(x= number_of_terrorist_attacks['Year'],y = number_of_terrorist_attacks['Terrorist attacks'], label = 'Terrorist Attacks')
sns.lineplot(x= number_of_terrorist_attacks['Year'],y = number_of_terrorist_attacks['Terrorism deaths'], label = 'Terrorist Deaths')
plt.xlabel('Year')
plt.ylabel('Number')
plt.title('Trends in Terrorist Attacks and Terrorism Deaths')
plt.legend()
plt.tight_layout()
plt.show()
```

## ***Insights***
Here are the key insights concisely:

- Declining trend in terrorist attacks and terrorism deaths over the years, with 2014 as the peak.
- Significant decrease in recent years, from 55,606 attacks and 140,534 deaths in 2016 to 33,207 attacks and 91,124 deaths in 2020.
- Potential factors influencing the trends, such as changes in security measures and counterterrorism efforts, require further analysis.
- Importance of continuous monitoring to understand the evolving nature of global terrorism and inform policy decisions.

### Which attack methods are most commonly used by terrorists?

```python
attacking_method = [col for col in df.columns if col.startswith('Attack method')]
attacking_method = df[attacking_method].sum().sort_values(ascending = False)
print(f'Most Commonly used attack methods by terrorists:\n {attacking_method}')
background_color = 'gray'
sns.set_theme(style="whitegrid", rc={"axes.facecolor": background_color})
plt.figure(figsize = (14,6))
plot = sns.barplot(x= attacking_method.index, y= attacking_method.values,edgecolor='white', width=0.8,palette="BrBG")
for container in plot.containers:
    plt.bar_label(container, label_type = 'edge', color = 'black',fontsize = 10,weight='bold',position=(1, 1),
              bbox = {'boxstyle' : 'round', 'facecolor' : 'white', 'edgecolor' : 'black'})
plt.xticks(rotation=90)
plt.title('Top Attack Methods Used by Terrorists',fontsize=16)
plt.xlabel('Attacking Method')
plt.ylabel('Frequency or Count')
plt.show()
```

## ***Insights***
Here are the key insights concisely:

- `Bombing/Explosion` is the most common attack method, with `380,697` incidents recorded.
- `Armed Assault` is the second most common attack method, with `87,655` incidents.
- `Assassination` is the third most common attack method, with `81,461` incidents.
- `Hostage Taking (Kidnapping)` is the fourth most common attack method, with `55,072` incidents.
- `Facility/Infrastructure` Attack is the fifth most common attack method, with `7,117` incidents.
- `Unarmed Assault` is the sixth most common attack method, with `4,535` incidents.
- `Hostage` Taking (Barricade Incident) is the seventh most common attack method, with `2,927` incidents.
- `Hijacking` is the least common attack method, with `2,927` incidents.

The data shows that explosive-related attacks (Bombing/Explosion) are the dominant attack method, followed by armed assaults and assassinations.

### Which countries or regions have experienced the highest frequency of terrorist attacks?

```python
attack_counts = df.groupby(['Entity'])['Terrorist attacks'].sum().sort_values(ascending = False).head(10)
print(f'Top 10 Entities by Frequency of Terrorist Attacks\n\n {attack_counts}')
# let's visualize
plt.figure(figsize = (14,6))
sns.set_theme(style="whitegrid", rc={"axes.facecolor": background_color})
plot = sns.barplot(x = attack_counts.index , y = attack_counts.values,edgecolor='white', width=0.8,palette="BrBG")
for container in plot.containers:
    plt.bar_label(container,label_type = 'edge',color = 'black',fontsize = 10,weight = 'bold',padding = 21,position=(1, 1),
                 bbox = {'boxstyle' : 'round', 'facecolor' : 'white', 'edgecolor' : 'black','pad' : 0.3, "alpha": 1})
plt.title('Top 10 Entities by Frequency of Terrorist Attacks',fontsize = 20)
plt.xlabel('Entities or Areas')
plt.ylabel('Number of Attacks')
plt.xticks(rotation = 45)
plt.show()
```

## ***Insights***
Here are the key insights concisely:

**Global Trends:** 
- The overall number of terrorist attacks worldwide is `201,272`, with the "World" entity accounting for the highest total.

**Regional Hotspots:**

- `Asia` is the region with the highest number of terrorist attacks at `121,548`.
- The `Middle East & North Africa` region is the second-highest with `59,514` attacks.
- South `Asia` is the third-highest with `56,578` attacks.

**Country-Level Analysis:**

- Within the broader regions, the country with the highest number of attacks is Afghanistan with 20,373 attacks.
- `Iraq` is the second-highest with `28,068` attacks.

**Geographical Spread:**

- `Africa` has experienced 32,843 attacks, with Sub-Saharan Africa accounting for `24,883` of those.
- `Europe` has seen `22,657` attacks.
- `South America` has experienced `19,993` attacks.
In summary, the data suggests that terrorism has had a significant global impact, with Asia, the Middle East & North Africa, and South Asia being the regions most affected. 

### Top 10 entity where terrorist death happes by doing suicide?

```python
suicide=df.groupby(['Entity'])['Terrorist Death Type : Suicide'].sum().reset_index()
suicide=suicide.sort_values(by="Terrorist Death Type : Suicide",ascending=False)
suicide.reset_index(drop=True, inplace=True) 
suicide=suicide.loc[1:10]
print(f'Terrorist death by doing suicide \n\n{suicide}')

plt.figure(figsize=(14, 8))
background_color = 'gray'
sns.set_theme(style="whitegrid", rc={"axes.facecolor": background_color})
plot = sns.barplot(y ="Entity",x="Terrorist Death Type : Suicide",data=suicide,edgecolor='white', width=0.8,palette="BrBG",orient = 'h')
for container in plot.containers:
    plt.bar_label(container,label_type = 'edge',color = 'black',fontsize = 10,weight = 'bold',padding = 21,position=(1, 1),
                 bbox = {'boxstyle' : 'round', 'facecolor' : 'white', 'edgecolor' : 'black','pad' : 0.3, "alpha": 1})
plt.title("Terrorist death by doing suicide",fontsize=15)
plt.ylabel("Entity",fontsize=15)
plt.xlabel("Terrorist Death Type : Suicide",fontsize=15)
ax = plt.gca()
ax.grid(True, axis='y', linestyle='-', alpha=0.3) 
plt.xticks(rotation=90) 
plt.tight_layout()
plt.show()
```

## ***Insights***
Here is a more concise analysis:

The data shows the highest terrorist deaths by suicide occurred in `Asia (59,292)`, followed by the `Middle East & North Africa (39,836)`. `Iraq (26,541)` and `South Asia (20,789)` also saw high numbers, driven by `Afghanistan (10,838)`. `Africa and Sub-Saharan Africa` had over `10,000` each, while `Pakistan (7,403)` and `Syria (6,524)` also saw significant suicide deaths.

### Total Death of different age people in different year?

```python
g1 =df.groupby(['Year'])['Terrorist attacks'].sum().reset_index()

g2 = df.groupby(['Year'])['Terrorism deaths'].sum().reset_index()

g3 = df.groupby(['Year'])['Death Age 100+'].sum().reset_index()

g4 = df.groupby(['Year'])['Death Age: 51-99 '].sum().reset_index()

g5 = df.groupby(['Year'])['Death Age : 21-50 '].sum().reset_index()

g6 = df.groupby(['Year'])['Death Age : 11-20 '].sum().reset_index()

g7 = df.groupby(['Year'])['Death Age : 6-10 '].sum().reset_index()

g8 = df.groupby(['Year'])['Death Age :  1-5'].sum().reset_index()
```

```python
fig, axes = plt.subplots(4, 2, figsize=(12, 15))
background_colors = 'white'
sns.set_theme(style="whitegrid", rc={"axes.facecolor": background_colors})

sns.lineplot(ax=axes[0, 0],data=g1, x='Year', y='Terrorist attacks')
axes[0, 0].set_title('Total Terrorist Attacks in Each Year')

sns.lineplot(ax=axes[0, 1],data=g2, x='Year', y='Terrorism deaths')
axes[0, 1].set_title('Total Death  in Each Year')

sns.lineplot(ax=axes[1, 0],data=g3, x='Year', y='Death Age 100+')
axes[1, 0].set_title(' Total  Death Age(100+) in Each Year')

sns.lineplot(ax=axes[1, 1],data=g4, x='Year', y='Death Age: 51-99 ')
axes[1, 1].set_title(' Total  Death Age(51-90) in Each Year')

sns.lineplot(ax=axes[2, 0],data=g5, x='Year', y='Death Age : 21-50 ')
axes[2, 0].set_title(' Total  Death Age(21-50) in Each Year')

sns.lineplot(ax=axes[2, 1],data=g6, x='Year', y='Death Age : 11-20 ')
axes[2, 1].set_title(' Total  Death Age(11-20) in Each Year')

sns.lineplot(ax=axes[3, 0],data=g7, x='Year', y='Death Age : 6-10 ')
axes[3, 0].set_title(' Total  Death Age(6-10) in Each Year')

sns.lineplot(ax=axes[3, 1],data=g8, x='Year', y='Death Age :  1-5')
axes[3, 1].set_title(' Total  Death Age(1-5) in Each Year')

plt.tight_layout()
plt.show()
```
